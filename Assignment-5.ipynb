## Define Visualize Function

```python
def visualize(filename):

    data = numpy.loadtxt(fname=filename, delimiter=',')

    fig = matplotlib.pyplot.figure(figsize=(12.0, 3.0))

#Setting up the axes
    axes1 = fig.add_subplot(1, 3, 1)
    axes2 = fig.add_subplot(1, 3, 1)
    axes3 = fig.add_subplot(1, 3, 1)

#Plotting average line
    axes1.set_ylabel('average')
    axes1.set_xlabel('Feature#')
    axes1.plot(numpy.mean(data, axis=0))

#Plotting maximum line
    axes2.set_ylabel('max')
    axes2.set_xlabel('Feature#')
    axes2.plot(numpy.max(data, axis=0))

#Plotting minimum line
    axes3.set_ylabel('min')
    axes3.set_xlabel('Feature#')
    axes3.plot(numpy.min(data, axis=0))       

#Graph settings
    fig.tight_layout()
    matplotlib.pyplot.grid(axis = 'y')
    matplotlib.pyplot.title(filename)
    matplotlib.pyplot.legend(['Avg','Max','Min'],loc="center right",fontsize=8)
    matplotlib.pyplot.ylabel("Values")
    matplotlib.pyplot.show()

#Remove warnings
import warnings
warnings.filterwarnings("ignore")
```

## Define Identify Issues Function

```python
def identify_issues(filename):

    data = numpy.loadtxt(fname=filename, delimiter=',')

#Add results to be printed using If and Elif    
    if numpy.max(data, axis=0)[0] == 0 and numpy.max(data, axis=0)[20] == 20:
        print('I doubt the integrity of this dataset\n')
    elif sum(numpy.min(data, axis=0)) == 0:
        print('Why is this happening?\n')
    else:
        print('Dataset looks good\n')
```

## Run the functions on the datasets

```python
#Import Libraries
import glob
import numpy 
import matplotlib
import matplotlib.pyplot 

#Run functions on the datasets
dataset = sorted(glob.glob('data*.csv'))
for datasets in dataset[:4]:
    print(datasets)
    visualize(datasets)
    identify_issues(datasets)
```
